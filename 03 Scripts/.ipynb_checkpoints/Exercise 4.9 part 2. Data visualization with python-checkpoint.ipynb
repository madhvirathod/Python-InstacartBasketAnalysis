## 3. Provide the Instacart senior stakeholders with descriptive findings about sales. create a histogram of the 'order_hour_of_day' column

```python
hist = instacart['order_hour_of_day'].plot.hist(bins = 24)
```

### This histogram provides a clear visual representation of customer ordering patterns throughout the day. It is evident that majority of orders are placed between 8 AM and 8 PM, with a significant peak around the mid-day hours.The frequency of orders rises sharply from 8 AM, reaches a maximum between 10 AM and 4 PM, and then gradually decreases as the evening progresses. This suggests that customers are most active around lunchtime.

## 5. Check whether there is a difference in expenditure(the 'prices' column) depending on the hour of the day

### Create a sample data.  70% in big and 30% in small.

```python
np.random.seed(4)
```

```python
dev = np.random.rand(len(instacart)) <= 0.7
```

```python
big = instacart[dev]
small = instacart[~dev]
```

```python
len(instacart)
```

```python
len(big) + len(small)
```

### Reduce the samples to only the columns needed.

```python
instacart_2 = small [['order_hour_of_day','prices']]
```

### lets check the subset

```python
instacart_2.head(5)
```

## Create a line chart

```python
line = sns.lineplot(data = instacart_2, x = 'order_hour_of_day',y = 'prices')
```

### The line chart shows that while there are some fluctuations in prices throughout the day, the average price trends to be more stable during regular shopping hours(morning to evening). The early morning and late evening hours exhibit more variability in average prices, which could be due to a smaller number of orders or a diverse range of items being ordered at those times.

## 7. Explore whether there is a connection between age and spending power(income) using a scatterplot

### Reduce previously made sample to only the columns needed

```python
instacart_4 = small [['age','income']]
```

### lets check the subset of the sample

```python
instacart_4.head(5)
```

## Create a scatterplot

```python
scatterplot = sns.scatterplot(x = 'age', y = 'income',data = instacart_4)
```

### The scaterplot shows the relationship between age and income. Each dot represents an individual. As the age increases, there is a wider range of incomes observed. The younger individuals (20-30) mostly have incomes clustered below 200,000 while older individuals(40 and above) have incomes spread across a broader range, up to 600,000. But there is no clear upward trend indicating that income increases consistently with age, the scatterplot shows that high-income individuals are more likely to be older. However, there are still many older individuals with lower incomes.
