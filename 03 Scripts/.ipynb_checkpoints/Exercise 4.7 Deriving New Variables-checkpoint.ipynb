## 1. create 'Price_label' column.

### create a subset of first million rows

```python
df = df_ords_prods_merge[:1000000]
```

```python
df.shape
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

### Use loc() method on df subset to apply price range function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

### Use loc() method on entire ord_prods_merge dataframe to apply price range function

```python
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['prices'] <= 15) & (df_ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

## Create 'busiest_day' column

```python
df_ords_prods_merge['order_day_of_week'].value_counts(dropna = False)
```

### Use For-loop to determine whether are on 'busiest day' (0=saturday), 'least busy' (4=Wednesday), or 'regular busy' (other days of week)

```python
result = []

for value in df_ords_prods_merge["order_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

### Create new column in ords_prods_merge to view results in context.

```python
df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

## Task: step 2 Update 'busiest day' column to 'busiest days' to identify order_day_of _week as 'busiest days' , 'slowest days' , or 'regularly busy'

### create column 'busiest_days' from 'order_day_of_week' to identify two busiest days(0,1), slowest days(4,3), or regularly busy(all other days).

```python
result_2 = []

for value in df_ords_prods_merge["order_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```

```python
result_2
```

## step 3 Check values of new 'busiest days' column for accuracy with obsevations in markdown format

```python
df_ords_prods_merge['busiest_days'] = result_2
```

## step 4 Create new column 'busiest_period_of_day' to identify time periods 'Most orders' , 'Average orders' and ' Fewest orders'

```python
df_ords_prods_merge['order_hour_of_day'].value_counts()
```

### The value counts listed above will be split into equal third for the following labels: Most orders = 10,11,14,13,15,12,16,9    Fewest orders = 23,6,0,1,5,2,4,3      Average orders = 17,8,18,19,20,7,21,22

### Create  For loop if statement labeling periods of time as 'Most orders', 'Average orders, and 'Fewest orders'

```python
result_3 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_3.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```

```python
result_3
```

### Create new column in 'busiest_period_of_day' in ords_prods_merge

```python
df_ords_prods_merge['busiest_period_of_day'] = result_3
```
